## 07: Vector Search with sqlitesearch

Key takeways:
1. sqlitesearch create a db file (writes onto disk), which can be easily opened/closed

```python
from sentence_transformers import SentenceTransformer
from sqlitesearch import VectorSearchIndex

model = SentenceTransformer("all-MiniLM-L6-v2")

vs_index = VectorSearchIndex(
    keyword_fields=["course"],
    mode="ivf",
    db_path="faq_vectors2.db"
)
```

```python
from rag_helper import RAGBase, RAGVector
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()
anthropic_client = Anthropic()
```

```python
vector_assistant = RAGVector(
  embedder=model,
  index=vs_index,
  llm_client=anthropic_client
)
```

```python
query = vector_assistant.rag("the program has already begun, can I still sign up?")
print(query)
```

```python
vs_index.close()
```
