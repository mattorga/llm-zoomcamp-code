```python
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()
client = Anthropic()
```

```python
prompt="Hello"
response = client.messages.create(
  model="claude-haiku-4-5",
  messages=[
    {
        "role": "user",
        "content": "Hello, Claude",
    }
  ],
  max_tokens=1024,
)

print(response.content[0].text)
```
